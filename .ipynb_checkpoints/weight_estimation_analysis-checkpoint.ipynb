```python
from ar import *
from vb import *
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity
```

```python
def estimate_output_density(samples,weight_index,true_value):
    samples = np.array(samples)[:, np.newaxis]
    X_plot = np.linspace(-5, 10, 1000)[:, np.newaxis]
    fig, ax = plt.subplots(figsize=(6,4))
    plt.title('Estimates of Weight #{0}  (True Value = {1})'.format(weight_index,true_value))
    kde = KernelDensity(kernel='gaussian', bandwidth=0.1).fit(samples)
    log_dens = kde.score_samples(X_plot)
    ax.plot(X_plot[:, 0], np.exp(log_dens), '-',label="Estimate Density")
    ax.legend(loc='upper left')
    ax.set_xlim(true_value - .5, true_value + .5)
    #ax.set_ylim(-0.02, 0.4)
    ax.plot(samples[:, 0], -0.005 - 0.01 * np.random.random(samples.shape[0]), '+k')
    plt.show()
```

```python
obs = [[],[],[],[]]
True_W = [.2, 0, -.4, -.2]
for t in range(25):
    # Time series is 3000 observations, burned in 500 observations.
    # The noise is a standard normal (Variance = 1)
    ts = generate_ar(3000,True_W,1,500)
    # Assume we know the true order
    # Our updates take evidence 50 observations at a time.
    W, updates, F = test_vb_ar_fit(ts,len(True_W),50)
    W = np.squeeze(np.asarray(W))
    for i in range(len(True_W)):
        obs[i].append(W[i])
```

```python
Weight_Number = 1
estimate_output_density(obs[Weight_Number-1],Weight_Number,True_W[Weight_Number-1])
```
